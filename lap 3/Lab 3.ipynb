# Step 4: Convert an image to different color spaces
If youhave  already read an image and want to convert it into GraySceel. You can use th code belowe.


```python
img = cv2.imread('./images/input.jpg')
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow('Grayscale image', gray_img)
cv2.waitKey()
```

Considering all the color spaces, there are around 190 conversation options available in OpenCV. If you want to see a list of all available flags,

You can convert to YUV by using the following flag: (YUV is another color model used to display color for Televisions.
- Y: physical linear-space brightness
- U: blue projection
- V: red projection

```python
yuv_img = cv2.cvtColor(img, cv2.COLOR_BGR2YUV)

# To separate out the three channels, write the following lines:
cv2.imshow('Y channel', yuv_img[:, :, 0]) 
cv2.imshow('U channel', yuv_img[:, :, 1])
cv2.imshow('V channel', yuv_img[:, :, 2])
cv2.waitKey()
```

# Task 1:

### Task 1.1:
Increase the size of the image. 

```python
from PIL import Image
img = Image.open('images/input.jpg')
width, height = img.size
resized = img.resize((width*2, height*2))
resized.show()
```

### Task 1.2:
Rotate the image by 120 degrees.

```python
rotated = img.rotate(120)
rotated.show()
```

### Task 1.3:
Perform Sheer operations over the image.

```python
sheered = img.transform(img.size, Image.AFFINE, (1, 0.3, 0, 0, 1, 0))
sheered.show()
```

Use any image processing library and apply the following intensity transformations over the image.

### Task 2.1: Negative

```python
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('images/input.jpg', 0) 
inverted = 255 - image

cv2.imshow("Inverted", inverted)
cv2.waitKey()
```

### Task 2.1: Log with (choose an appropriate constant for your image)

```python
c = 255 / np.log(1 + np.max(image))
log_transformed = c * np.log(1 + image.astype(np.float64))
log_transformed = np.array(log_transformed, dtype=np.uint8)

cv2.imshow("Log Tranformed", log_transformed)
cv2.waitKey()
```

### Task 2.1: Power law (choose a value for gamma that increases the contrast)

```python
gamma = 2.2
power_tranformed = np.array(255*(image / 255) ** gamma, dtype = 'uint8')

cv2.imshow("Power Tranformation", power_tranformed)
cv2.waitKey()
```
